```python
%matplotlib inline
import pandas as pd
from fbprophet import Prophet

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
years = ["2008-01-01", "2009-01-01", "2010-01-01", "2011-01-01", "2012-01-01", "2013-01-01", "2014-01-01", "2015-01-01", "2016-01-01"]
scores = [212,323,377,338,343,349,329,343,337]
df = pd.DataFrame({'y': scores, 'ds': years})
df.dtypes
df.head()
```

```python
df.dtypes
```

```python
df['ds'] = pd.DatetimeIndex(df['ds'])
df.dtypes
```

```python
ax = df.set_index('ds').plot(figsize=(12,8))
ax.set_ylabel('Scores')
ax.set_xlabel('Year')

plt.show()
```

```python
my_model = Prophet(interval_width=0.95)
```

```python
my_model.fit(df)
```

```python
future_dates = my_model.make_future_dataframe(periods=1, freq="YS")
future_dates.tail()
```

```python
forecast = my_model.predict(future_dates)
forecast.tail()
```

```python
my_model.plot(forecast, uncertainty=True)
```

```python
my_model.plot_components(forecast)
```

```python
m = Prophet()
m.add_seasonality(name='yearly', period=365, fourier_order=3, prior_scale=0.1)
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```
